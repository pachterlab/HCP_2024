## Figure 1C: barplots.

```python
def plot_fig1_barplots(height1,height2):
  fig, axs = plt.subplots(2,1,figsize=(10,10))

  ax = axs[0]
  ax.bar(x = ['$X_1$','$X_2$','','$Y_1$','$Y_2$'],
       height=height1,
      color=['magenta','gray','white','magenta','gray'],
      alpha = 0.5)
  ax.spines['top'].set_visible(False)
  ax.spines['right'].set_visible(False)
  ax.tick_params(axis='both', which='both',
               bottom=False, top=False, left=False, right=False,
               labelbottom=False, labelleft=False)

  ax = axs[1]
  ax.bar(x = ['$X_1$','$X_2$','','$Y_1$','$Y_2$'],
       height=height2,
      color=['magenta','gray','white','magenta','gray'],
      alpha = 0.5)
  ax.spines['top'].set_visible(False)
  ax.spines['right'].set_visible(False)
  ax.tick_params(axis='both', which='both',
               bottom=False, top=False, left=False, right=False,
               labelbottom=False, labelleft=False)

  plt.subplots_adjust(hspace=0.4)

  plt.show()
```

```python

# cis
scale = 1
p1 = 5
p2 = 2.5
p3 = 5 * scale
p4 = p3*(p2/p1) * scale

height1 = [p1,p2,0,p3,p4]
height2 = [p2,p1,0,p4,p3]
plot_fig1_barplots(height1,height2)
```

```python

# cis
scale = 1
p1 = 5
p2 = 2.5
p3 = 5 * scale
p4 = p3*(p2/p1) * scale

height1 = [p1,p2,0,p3,p4]
height2 = [p2,p1,0,p4,p3]
plot_fig1_barplots(height1,height2)
```

```python
# cis + trans
p1 = 5
p2 = 2.5
p3 = 5
p4 = 4.0

height1 = [p1,p2,0,p3,p4]
height2 = [p2,p1,0,p4,p3]
plot_fig1_barplots(height1,height2)
```

```python
# cis x trans
p1 = 5
p2 = 2.5
p3 = 5
p4 = 10


height1 = [p1,p2,0,p3,p4]
height2 = [p2,p1,0,p4,p3]
plot_fig1_barplots(height1,height2)
```

## Figure 1D: theoretical geometry/diagram.

```python
fig, axs = plt.subplots(1,3, figsize = (18.6,6) )

# Data
x = np.linspace(-5, 5, 100)

axis_width = 10
alpha = 0.25
line_alpha = 0.7
fill_alpha = 0.4


ax = axs[0]
ax.plot(x, x, color='orangered', linewidth=axis_width,alpha = line_alpha)
ax.axhline(0, color='royalblue', linewidth=axis_width, alpha=line_alpha)
# ax.plot(x+x, x, color='skyblue', linewidth=axis_width,alpha = line_alpha)
# ax.plot(-x-x, x, color='forestgreen', linewidth=axis_width,alpha = line_alpha)
ax.axhline(0, color='black', linewidth=2, alpha=0.7, linestyle='-')
ax.axvline(0, color='black', linewidth=2, alpha=0.7, linestyle='-')

# fill between
ax.fill_between(x=[-5,0], y1=0, y2=5, color='forestgreen', alpha = fill_alpha)
ax.fill_between(x=[0,5], y1=0, y2=-5, color='forestgreen', alpha = fill_alpha)
ax.fill_between(x=[-5,0], y1=-5, y2=[-5,0], color='forestgreen', alpha = fill_alpha)
ax.fill_between(x=[0,5], y1=5, y2=[0,5], color='forestgreen', alpha = fill_alpha)
ax.fill_between(x=[-5,0], y1=0, y2=[-5,0], color='skyblue', alpha = fill_alpha)
ax.fill_between(x=[0,5], y1=0, y2=[0,5], color='skyblue', alpha = fill_alpha)

# Set limits and turn off spines
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.tick_params(axis='both', which='both', bottom=False, top=False, left=False, right=False, labelbottom=False, labelleft=False)


ax = axs[1]
ax.axvline(0, color='orangered', linewidth=axis_width, alpha=line_alpha)
ax.axhline(0, color='royalblue', linewidth=axis_width, alpha = line_alpha)
# ax.plot(x, x, color='skyblue', linewidth=axis_width,alpha = line_alpha)
# ax.plot(-x, x, color='forestgreen', linewidth=axis_width,alpha = line_alpha)
ax.axhline(0, color='black', linewidth=2, alpha=0.7, linestyle='-')
ax.axvline(0, color='black', linewidth=2, alpha=0.7, linestyle='-')

# fill between
ax.fill_between(x=[-5,0], y1=0, y2=5, color='forestgreen', alpha = fill_alpha)
ax.fill_between(x=[0,5], y1=0, y2=-5, color='forestgreen', alpha = fill_alpha)
ax.fill_between(x=[-5,0], y1=0, y2=-5, color='skyblue', alpha = fill_alpha)
ax.fill_between(x=[0,5], y1=0, y2=5, color='skyblue', alpha = fill_alpha)


# Set limits and turn off spines
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.tick_params(axis='both', which='both', bottom=False, top=False, left=False, right=False, labelbottom=False, labelleft=False)



ax = axs[2]
ax.axvline(0.5, color='orangered', linewidth=axis_width, alpha=line_alpha)
ax.axvline(-0.5, color='royalblue', linewidth=axis_width, alpha=line_alpha)
ax.axvline(0.0, color='skyblue', linewidth=axis_width, alpha=line_alpha)
ax.axvline(-1, color='forestgreen', linewidth=axis_width, alpha=line_alpha)
ax.axvline(1, color='forestgreen', linewidth=axis_width, alpha=line_alpha)
ax.axhline(0, color='gray', linewidth=axis_width, alpha = line_alpha)
ax.axhline(0, color='black', linewidth=2, alpha=0.7, linestyle='-')

# fill between
ax.fill_between(x=[-1,-0.5], y1=0, y2=1.0, color='forestgreen', alpha = fill_alpha)
ax.fill_between(x=[0.5,1], y1=0, y2=1.0, color='forestgreen', alpha = fill_alpha)
ax.fill_between(x=[-0.5,0.5], y1=0, y2=1.0, color='skyblue', alpha = fill_alpha)

# Set limits and turn off spines
ax.set_xlim(-1.02, 1.02)
ax.set_ylim(-0.01, 1)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.tick_params(axis='both', which='both', bottom=False, top=False, left=False, right=False, labelbottom=False, labelleft=False)

plt.subplots_adjust(wspace=0.2)

plt.savefig('./figs/fig1C',dpi=450,bbox_inches='tight')
plt.show();
```
